# Fig. 5: petitRADTRANS chem. eq. spectral templates of WASP-189 b

```python
import os
import numpy as np
from matplotlib import pyplot as plt
import pickle
from astropy.modeling import models
import astropy.units as u
```

```python
APJ_SINGLE_COL = 3.5
APJ_DOUBLE_COL = 7.
TEXTSIZE=12

tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),  
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),  
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),  
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),  
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]

# Scale to range of 0-1 for matplotlib
for i in range(len(tableau20)):  
    r, g, b = tableau20[i]  
    tableau20[i] = (r / 255., g / 255., b / 255.)
```

Plot figure.

```python
dirin = '../data/pRT_spectral_templates'

spectral_templates_fnames = np.array([
    'W189b_pRT_1D_Yan22_chem_eq_39K__Allard.txt',
    'W189b_pRT_1D_Yan22_chem_eq_23Na__Allard.txt',
    'W189b_pRT_1D_Yan22_chem_eq_Fe.txt',
    'W189b_pRT_1D_Yan22_chem_eq_FeH.txt',
    'W189b_pRT_1D_Yan22_chem_eq_Ti.txt',
    'W189b_pRT_1D_Yan22_chem_eq_Ti-O-NatAbund__Toto.txt',
    'W189b_pRT_1D_Yan22_chem_eq_V.txt',
    'W189b_pRT_1D_Yan22_chem_eq_51V-16O__HyVO.txt',
    'W189b_pRT_1D_Yan22_chem_eq_Ca.txt',
    'W189b_pRT_1D_Yan22_chem_eq_Ca_p.txt',
    'W189b_pRT_1D_Yan22_chem_eq_Si.txt',
    'W189b_pRT_1D_Yan22_chem_eq_Mg.txt',
    'W189b_pRT_1D_Yan22_chem_eq_Cr.txt',
    'W189b_pRT_1D_Yan22_chem_eq_Al.txt',
    'W189b_pRT_1D_Yan22_chem_eq_all.txt'
])[::-1]

specie_labels = np.array([
    'K',
    'Na',
    'Fe',
    'FeH',
    'Ti',
    'TiO',
    'V',
    'VO',
    'Ca',
    r'Ca$^+$',
    'Si',
    'Mg',
    'Cr',
    'Al',
    'Fiducial'
])[::-1]

colors = [
    tableau20[4],
    tableau20[4],
    tableau20[10],
    tableau20[10],
    tableau20[2],
    tableau20[2],
    tableau20[8],
    tableau20[8],
    tableau20[12],
    tableau20[12],
    tableau20[18],
    tableau20[16],
    tableau20[14],
    tableau20[14],
    'k'
][::-1]

spectral_templates = [
    np.loadtxt(os.path.join(dirin, f)) for f in spectral_templates_fnames
]

Rstar = (2.33 * u.Rsun).to(u.au).value
Rp = (1.619 * u.Rjup).to(u.au).value
area_ratio = (Rp/Rstar)**2

Tstar = 8000.
bbf = models.BlackBody(temperature=Tstar*u.K)



plt.figure(figsize=(APJ_SINGLE_COL, APJ_SINGLE_COL*1.668))
for n, st in enumerate(spectral_templates):
    if n == 0:
        c = 'k'
    else:
        c = tableau20[(n-1)*2%20]
    wl, flux = st
    x = wl/1e-4
    y = np.log10(flux/bbf(wl/1e-4 * (1e-6 * u.m)).value * area_ratio * 1e4) + n
    plt.plot(x, y, color=colors[n])
    plt.annotate(specie_labels[n], xy=(1.41, np.median(y)), color=colors[n], fontsize=TEXTSIZE)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.yticks(np.arange(0, len(spectral_templates)))
plt.ylabel(r'log$_{10}$ (planet-to-star contrast ratio $\times 10^4$ + offset)', fontsize=TEXTSIZE)
plt.xlim(1.1, 1.42)
plt.ylim(0,)
plt.xlabel(r'Wavelength [$\rm{\mu}$m]', fontsize=TEXTSIZE)
plt.title('petitRADTRANS chemical equilibrium\nspectral templates', fontsize=TEXTSIZE)

plt.gca().tick_params(axis ='both', labelsize=TEXTSIZE)

fname = 'Fig5_pRT_spectral_templates_chem_eq'
plt.savefig('../figures/'+fname+'.png', dpi=200, bbox_inches='tight')
plt.savefig('../figures/'+fname+'.pdf', bbox_inches='tight')
plt.show()
```
